```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("insurance_data.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x='age',y='bought_insurance',data=df,marker='x',color='Green')
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['age']],df.bought_insurance,train_size=0.9,random_state=10)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)# 1 means he will buy insurance and vice-versa
```

```python
model.score(x_test,y_test)*100
```

```python
model.predict_proba(x_test)
```
