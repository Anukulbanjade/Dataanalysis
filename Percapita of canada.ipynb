```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn import linear_model
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Dataset/canada_per_capita_income.csv')
```

```python
sns.scatterplot(x='year',y='per capita income (US$)',data=df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
year_to_predict = 2020
predicted_income = reg.predict([[year_to_predict]])
print(f"Predicted per capita income for {year_to_predict}: ${predicted_income[0]:.2f}")
```

```python

sns.regplot(x='year', y='per capita income (US$)', data=df, color='black')
```

```python
reg.score(df[['year']],df['per capita income (US$)'])
```
