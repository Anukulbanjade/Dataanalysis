```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:/Users/anuku/Downloads/homeprices.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
sns.scatterplot(x='area',y='price',data=df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
value_to_predict = np.array([5000]).reshape(-1, 1)
```

```python
predicted_price = reg.predict(value_to_predict)
print(predicted_price)
```

## Mathematiacal Operation behind Regression model

```python
m=reg.coef_#m
```

```python
b=reg.intercept_#b
```

```python
y=m*5000+b
print(y)
```
